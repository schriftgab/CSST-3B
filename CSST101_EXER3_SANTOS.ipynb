# **Posterior Probability:**

```python

P_Hypertension = 0.15
P_No_Hypertension = 1 - P_Hypertension


P_Headaches_given_Hypertension = 0.6
P_Shortness_of_Breath_given_Hypertension = 0.5
P_Nosebleeds_given_Hypertension = 0.3


P_Headaches_given_No_Hypertension = 0.1
P_Shortness_of_Breath_given_No_Hypertension = 0.05
P_Nosebleeds_given_No_Hypertension = 0.02


P_Symptoms_given_Hypertension = (P_Headaches_given_Hypertension *
                                  P_Shortness_of_Breath_given_Hypertension *
                                  P_Nosebleeds_given_Hypertension)

P_Symptoms_given_No_Hypertension = (P_Headaches_given_No_Hypertension *
                                     P_Shortness_of_Breath_given_No_Hypertension *
                                     P_Nosebleeds_given_No_Hypertension)

P_Symptoms = (P_Symptoms_given_Hypertension * P_Hypertension +
              P_Symptoms_given_No_Hypertension * P_No_Hypertension)


P_Hypertension_given_Symptoms = (P_Symptoms_given_Hypertension * P_Hypertension) / P_Symptoms


print(f"Probability of having Hypertension given symptoms: {P_Hypertension_given_Symptoms:.4f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# Parameters for the binomial distribution
n = 5  # Number of trials (coin flips)
p = 0.5  # Probability of getting heads

# Create an array of possible outcomes (0 to n heads)
x = np.arange(0, n + 1)

# Calculate the probability mass function (PMF) for each outcome
pmf = binom.pmf(x, n, p)

# Plotting the probability distribution
plt.bar(x, pmf, color='skyblue')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.title('Probability Distribution of Getting 0 to 5 Heads in 5 Coin Flips')
plt.xticks(x)  # Show all possible outcomes on x-axis
plt.ylim(0, max(pmf) + 0.05)  # Set y-axis limits slightly above max probability
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
